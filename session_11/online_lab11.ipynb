We are going to review the following knowledge points:

1. How to create simple plots in Pandas
2. How to change axis limits and Figure size
3. How to move legends, format axes and create multiple plots


#### Ex1:  Read in the mtcars dataset and use style ggplot

```python
%matplotlib inline

#Import here
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use("ggplot")

df_mtcars = pd.read_csv("data/mtcars.csv")
df_mtcars.head()
```

#### Ex2:  Plot a scatter plot of mpg and number of cylinders for all car models. The x-axis should be named as "Miles per gallon", while the y-axis should be named as "Number of cylinders". 


```python
fig, ax = plt.subplots()

df_mtcars.plot(kind = "scatter", x="mpg", y="cyl", ax = ax)
ax.set(xlabel = "Miles per Gallon", ylabel = "Number of cylinders")
```

#### Ex3:  First, create a figure with two subplots in one row. Those subplots should share the same y-axis. The names for the entire figure is "Different Cars MPG and Horsepower". 

```python
fig, (ax0, ax1) = plt.subplots(nrows = 1, ncols = 2, figsize = (10,8), sharey = True)
fig.suptitle("Different Cars MPG and Horsepower")
```

#### Ex4: Generate a horizontal bar chart of the mileage per gallon for all cars. The x-axis should be named as "Mileage per gallon", while the y-axis should be "Car make". The bar chart should be in red. Set the limit of x-axis between 5 and 35. 

```python
df_mtcars.plot(kind = "barh", x = "car_name", y = "mpg", color = "r", ax=ax0)
ax0.set(title = "Mileage per Gallon", xlabel = "Mileage per Gallon", ylabel = "Car Make")
ax0.set_xlim([5, 35])

fig
```

#### Ex5: Add a vertical dashed blue line at the average mpg. Add text at the right side of line saying "Average MPG = xxx" xxx is at 2-decimal point precision.

```python
mean_mpg = df_mtcars.mpg.mean()
ax0.axvline(mean_mpg, color = "b", linestyle = "--")
text = "Average MPG = %0.2f" % mean_mpg
ax0.annotate(text, xy = (mean_mpg+1, 15))

fig
```

#### Ex6:  Generate the right-hand side bar chart which contains the horsepower for each car. There should also be a vertical blue line on the average horsepower and text annotation. The bars should be green and x-axis should be between 40 to 350. 

```python
df_mtcars.plot(kind = "barh", x="car_name", y="hp", color = "g", ax=ax1)
ax1.set(title="Horsepower", xlabel = "Horsepower")
ax1.set_xlim([40,350])

mean_hp = df_mtcars.hp.mean()
ax1.axvline(mean_hp, color = "b", linestyle = "--")
text = "Average HP %0.2f" % mean_hp
ax1.annotate(text, xy=(mean_hp+10, 15))

fig
```
